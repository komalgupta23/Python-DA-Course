```python
import pandas as pd
```

```python
df = pd.read_excel('Canada.xlsx',sheet_name = 1, skiprows = 20, skipfooter = 2)
df
```

```python
df.columns
```

```python
df['Type']
```

```python
df[['Type', 1980, 2000, 'AreaName']] # selecting multiple columns as a list
```

```python
cols = ['Type', 'Coverage', 'AREA', 'REG', 'DEV']
df[cols]
```

```python
# preprocessing the data
df = df.drop(columns = cols)
df.head(2)
```

```python
df = df.rename(columns = {
        'OdName' : 'Country',
        'AreaName' : 'Continent',
        'RegName' : 'Region',
        'DevName' : 'Status'
    })
df.head(2)
```

```python
years = list(range(1980, 2014))
df['total'] = df[years].sum(axis = 1)
df.head()
```

```python
df = df.sort_values(by = 'total')
df.tail()
```

```python
df = df.set_index('Country')
df.head(5)
```

```python
df.loc['Japan', years].mean()
```

```python
df.loc['Australia', years].mean()
```

```python
import plotly.express as px
```

```python
japan_df = df.loc['Japan',years]
fig = px.line(japan_df,
              x = japan_df.index,
              y= japan_df.values,
              text = japan_df.values,)
fig.update_traces(textposition = 'top center')
fig
```

```python
country_df = df.loc[['Japan','Australia'], years].T
country_df
```

```python
countries = ['Japan','Australia']
country_df = df.loc[countries, years].T
px.line(country_df,
        x=country_df.index,
        y=['Japan','Australia'],)
```

get a single group from dataframe

```python
df.groupby('Continent').get_group('Europe')
```

```python
# similar to groupby but can be used for multiple conditions
df[df['Continent'] == 'Europe']
```

statistical operation on grouped data

```python
# groupby(cat_col)[num_col(s)].operation()
df.groupby('Continent')['total'].sum()
```

```python
continent_immigration = df.groupby('Continent')['total'].sum()
px.pie(continent_immigration, names = continent_immigration.index, values = continent_immigration.values, hole = 0.5)
```

```python
con_in_trend = df.groupby('Continent')[years].sum()
con_in_trend.T
```

```python
px.line(con_in_trend.T, 
        x = con_in_trend.columns, 
        y = con_in_trend.index, 
        title = 'Immigration trend by continent', 
        labels = {'x' : 'Years', 'y' : 'Number'}
        )
```

pivot table

```python
df.pivot_table(index = 'Continent', columns = 'Status', values = 'total', aggfunc = 'sum')
```

```python
status_df = df.pivot_table(index = 'Region', columns = 'Status', values = 'total', aggfunc = 'sum')
status_df
```

```python
px.bar(status_df,
       x = status_df.index,
       y = status_df.columns,
       title = 'Immigration by region and status')
```

```python
df['80s'] = df[range(1980, 1990)].sum(axis = 1)
df['90s'] = df[range(1990, 2000)].sum(axis = 1)
df['2k'] = df[range(2000, 2010)].sum(axis = 1)
df['2k_rem'] = df[range(2010, 2014)].sum(axis = 1)
df.head()
```

```python
decade_df = df[['80s', '90s', '2k', '2k_rem']]
decade_df = decade_df.sort_values(by = ['80s', '90s', '2k', '2k_rem'],
                                  ascending = False).head(10)
px.bar(decade_df,y = decade_df.index,
       x = decade_df.columns,)
```
