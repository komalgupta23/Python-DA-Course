JOIN, MERGE AND CONCATENATE IN PANDAS

```python
data1 = {"Emp ID":["E01","E02", "E03", "E07", "E05"],
         "Name":["Ram", "Shyam", "Komal", "Sahil", "Rahul"],
         "Age":[23, 45, 34, 35, 32]}

data2 = {"Emp ID":["E01","E06", "E03", "E04", "E05"],
         "Salary":[35000, 45000, 54000, 58000, 40000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df1, df2
```

```python
# merge

pd.merge(df1, df2 , on = "Emp ID")
```

```python
# join

pd.merge(left = df1 , right = df2 , how = "left")
```

```python
pd.merge(left = df1 , right = df2 , how = "right")
```

```python
# concatenate

pd.concat([df1, df2])
```

COMPARE DATAFRAME

```python
dict = {"Fruits":["Mango", "Banana", "Apple", " Papaya"],
        "Price": [100, 150, 50, 35],
        "Quantity":[15, 10, 10, 3]}

f1 = pd.DataFrame(dict)

f2 = f1.copy()

f2.loc[0, "Price"]= 120
f2.loc[1, "Price"]= 170
f2.loc[3, "Price"]= 30
f2.loc[0, "Quantity"]= 30
f2.loc[1, "Quantity"]= 15
f2.loc[3, "Quantity"]= 5

#f1 ,f2

f1.compare(f2)
```
